```python
import matplotlib.pyplot as plt
import numpy as np

def gen_qpsk(B, samples_per_bit, w):
    bits = np.random.randint(low=0, high=3, size=B)
    bits_ups = np.repeat(bits, samples_per_bit)
    
    n= np.arange(len(bits_ups))    
    s = np.exp(1j*(np.pi/2)*bits_ups)*np.exp(1j*2*np.pi*w*n)

    return s

s = gen_qpsk(10, 10, 0.05)

plt.plot(np.real(s))
plt.plot(np.imag(s))

s4 = s**4

plt.figure()
plt.plot(np.real(s4))
plt.plot(np.imag(s4))

S=np.fft.fft(s)
Sxx=np.abs(S)**2

S4=np.fft.fft(s4)
S4xx=np.abs(S4)**2

plt.figure()
plt.plot(Sxx)
plt.plot(S4xx)
```

```python
w_arr = [-0.1, 0.2, 0.3]
B=1000
spb = 300

s = np.zeros(B*spb)

for i in range(len(w_arr)):
    sig = gen_qpsk(B, spb, w_arr[i])
    s = s + sig

plt.plot(np.real(sig))

S = np.fft.fft(s)
S_psd = np.abs(S)**2

plt.figure()
plt.plot(np.fft.fftshift(S_psd))

s4 = s**4
S4 = np.fft.fft(s4)
S4_psd = np.abs(S4)**2

plt.figure()
plt.plot(np.fft.fftshift(S4_psd/100))
```
